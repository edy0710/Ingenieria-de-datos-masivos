```python
# Incluya aquí su código para importar librerías y datos
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Cargar los datos desde el archivo CSV
data = pd.read_csv("/content/sample_data/responses.csv")
```

```python
# Verificar datos
print(data.head())
print(data.info())
```

```python
# Código para crear un dataframe solo con los campos de variables Música y Demografía
# Seleccionamos solo columnas numéricas
numeric_cols = data.select_dtypes(include=[np.number]).columns.tolist()

# Filtramos para mantener solo las de música (primeras 19) y demografía (últimas 10)
variables_musica = numeric_cols[:19]  # Primeras 19 columnas numéricas (música)
variables_demografia = numeric_cols[-10:]  # Últimas 10 columnas numéricas (demografía)
lista_variables = variables_musica + variables_demografia

data_mus_dem = data[lista_variables]
```

```python
# Calcular matriz de correlación
corr_matrix = data_mus_dem.corr()
```

```python
# Visualización de la matriz de correlación
plt.figure(figsize=(20, 15))
sns.heatmap(corr_matrix,
            cmap='coolwarm',
            center=0,
            annot=True,
            fmt=".2f",
            linewidths=.5,
            cbar_kws={"shrink": .8})
plt.title('Matriz de Correlación entre Variables de Música y Demografía', pad=20, fontsize=16)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# Obtener las 10 correlaciones más altas (positivas y negativas)
corr_unstacked = corr_matrix.unstack().sort_values(ascending=False)
corr_pairs = corr_unstacked[corr_unstacked != 1].drop_duplicates()
```

```python
print("\n10 correlaciones más fuertes (positivas y negativas):")
print(corr_pairs.head(10))  # Top 5 positivas
print(corr_pairs.tail(10))  # Top 5 negativas
```
